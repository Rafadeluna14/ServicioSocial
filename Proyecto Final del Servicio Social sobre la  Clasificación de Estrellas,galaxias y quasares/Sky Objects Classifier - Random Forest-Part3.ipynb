```python
featuredf = df.drop(['class','objid', 'run', 'rerun', 'camcol', 'field', 'specobjid', 'plate', 'mjd', 'fiberid', 'redshift'], axis=1)

featurecols = featuredf.keys()
astrObjs = df['class'].unique()

plt.figure(figsize=(15,10))
for i in range(len(featurecols)):
    plt.subplot(2, 4, i+1)
    for j in range(len(astrObjs)):
        sns.distplot(df[df['class'] == astrObjs[j]][featurecols[i]], 
                     hist = False, kde = True, 
                     kde_kws = {'shade': True, 'linewidth': 1}, 
                     label = astrObjs[j])
    plt.legend()
    plt.xlabel(featurecols[i])
plt.tight_layout()
```
